**Introduction**


The following is a literature critique of the research presented by Belhumeur, Hespanha and Kriegman and their approaches toward developing a complex facial recognition algorithm whose outcomes are irrespective of both varying light directions and facial expressions.
By focusing on two methods [Eigenfaces and Fisherfaces] the authors aim to explain and justify why both methods are effective and yet still, imperfect given the biologically induced challenges that our bodies present that in turn, do not so easily comply with computational methods made for executing these underlying scientific truths.
In essence, this report looks to critique the authors’ approaches towards solving the digital challenge of facial recognition. 


**Content**


The paper in review discusses the challenges faced by three researchers in trying to develop a new and sophisticated facial recognition algorithm that is unperturbed by varying conditions.
The entire paper shows how the authors are able to utilise techniques of Linear Discriminant Analysis to decipher effective processes that can be used by systems to mathematically determine an individual from the image taken of them. 
The algorithms being compared are Eigenfaces, Fisherfaces, Correlation and Linear Subspaces. As it pertains to facial recognition, Belhumeur et. al (1997) states that most algorithms are reliant on the image of a face being treated as a ‘3D linear subspace’, that is, the face should become a mappable plane with multiple coordinates- so to speak. The issue with this is that, in practice, the digital mapping of a surface into coordinates can prove difficult due to a number of reasons. Reasons such as surface reflections, shadows, varying expressions, camera angles and changing facial features such as hair, accessories etc.

Firstly, the correlation method utilises the classifier of nearest neighbour on the image. The images in the test and training sets are normalised to have a mean of zero and unit variance of zero also.
From here, all unrecognised images are gathered together and are numerically analysed to discern which image looks closest to the test set [image]. 

The Eigenfaces algorith maximises the scatter of the entire set using principal components analysis [PCA]. Simply put, Eigenfaces projects as many lighting variations as possible to maximise the system's potential of recognising the image. This algorithm is computationally simpler to execute than correlation algorithm.


The next algorithm being explored is that of linear subspaces.
This algorithm requires that surfaces, in this case, the face, be treated as a Lambertian surface. That is, that the image[s] of a face can be mapped in a 3D linear subspace of the overarching image space.
Linear subspaces 'represent all possible images of a specific object under specific light confitions' (Elhabian 2012).
This algorithm is purposed for the digital reconstruction of each face using atleast three images 'taken under different lighting directions' (Belhumeur et. al 1997).

Fisherfaces is an algorithm built on the linear subspaces algorithm whilst using Fisher's Linear Discriminant. It capitalises on some of the same techniques as those of linear subspaces to accurately depict images and their closest matches.
This algorithm may usurp those previosuly mentioned as it maximises the ratio of the between and the within-class scatter (Belhumeur et. al 1997).

Two experiments were conducted, each of which compared the four algorithms discussed. The first one experimented with varied lighting whilst the other assessed both varied lighting and facial expressions.
One other area of innovation is in how they intentionally experiment with extremes of each photo in order to develop a better algorithm. Namely, in using the Eigenfaces algorithm, the authors sought to experiment with images each of which had to assume a 'ghastly appearance' (Lata et. al 2009) in order to train the data set to learn what to discern from imags of varying lighting conditions.
